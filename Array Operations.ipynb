```python
import numpy as np

array1=np.array([1,2,3,4,5])
array2=np.array([6,7,8,9,10])

#sum 2 arrays
sum = array1+array2
print(sum)
```

```python
subtract = array1-array2
print(subtract)
```

```python
multiply=array1*array2
print(multiply)
```

```python
div=array1/array2
print(div)
```

```python
mean=np.mean(array1)
print(mean)

mean2=np.mean(array2)
print(mean2)
```

```python
sum=np.sum(array1)
print(sum)

sum2=np.sum(array2)
print(sum2)
```

```python
max=np.max(array1)
max2=np.max(array2)
print("max array1: ",max,"\nmax array2: ",max2)

min=np.min(array1)
min2=np.min(array2)
print("min array1: ",min,"\nmin array2: ",min2)
```

```python
greater_than = array1>2
print("greater than: ",greater_than)

less_than = array1<2
print("less than: ",less_than)
```

```python
#adding num to all values in the array
result=array1+5
print(result)
```
