```python
#note: tuples are defined using () - immutable (can't be changed unless turned to list)

try:
  employee=(1,'Jane Doe', 'Cebu', 991432421)
  print(employee)
  print(employee[-1])
  print(len(employee))

except Exception as e:
  print ("error: ", e)
```

```python
try:
  lst_employee=list(employee)
  lst_employee[1]='Jane Doe Cutamora'
  employee=tuple(lst_employee)
  print(employee)

except Exception as e:
  print("error: ",e)
```

```python
for x in employee:
  print(x)
```

```python
#note: list are defined using [] - mutable (changeable)

tasks=['development','testing','monitoring']

try:
  print(tasks)
  tasks[1]='regression testing'
  print(tasks)
  #add item
  tasks.append('production deployment')
  print(tasks)
  #insert item from index
  tasks.insert(0,'requirements gathering')
  print(tasks)
  #remove item from list
  tasks.pop(3)
  print(tasks)

except Exception as e:
  print("error: ",e)
finally:
  for x in tasks:
    print(x)
```
