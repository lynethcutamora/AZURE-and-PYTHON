```python
#sets
#note: defined using {} - not ordered (no index) - no duplicates

try:
  tasks={'development','testing','monitoring'}
  print(tasks)
  #add new item
  tasks.add('deployment')
  #add items using list
  ls={'unit testing','regression testing'}
  tasks.update(ls)
  #remove item
  tasks.remove('testing')

  #clear ls
  ls.clear()
  print(ls)
except Exception as e:
  print("error: ",e)
finally:
  for x in tasks:
    print(x)
```

```python
#dictionaries
#defined using {} - ordered - goes in pair - no duplicates

import pandas as pd

try:
  user={"name": "Jane Doe","city": "Cebu","age": 30}
  print(user['age'])
  #using get
  print(user.get('name'))
  #print keys
  keys=user.keys()
  print(keys)
  #print values
  values=user.values()
  print(values)

  if 'fname' in user:
    print('true')
  else:
    print('false')

  #modify value
  user['age']=29
  print(user)

  #update dictionary
  user.update({'gender':'female'})
  print(user)

except Exception as e:
  print("error: "+e)
finally:
  print("-----------")
  for x in user.keys():
    print(user[x])
```
