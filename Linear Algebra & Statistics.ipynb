```python
#note: matrix - 2 dimensional data structure (contains numbers / objects for linear algebra)
import numpy as np

array1=np.array([[1,2],[3,4]])
array2=np.array([[5,6],[7,8]])

#multipy using dot
mul=np.dot(array1,array2)
print(mul)
```

```python
#inversion

invr=np.linalg.inv(array1)
print(invr)
```

```python
#evaluate eigenvalues and eigenvectors

eval,evec=np.linalg.eig(array1)

print("eigenvalues: ",eval,"\neigenvectors: ",evec)
```

```python
#standard deviation and mean

data=np.array([1,2,3,4,5])
mean=np.mean(data)

std_dev=np.std(data)

print("mean: ",mean,"\nstandard deviation: ",std_dev)
```

```python
#correlation coefficient

data2=np.array([11,12,13,14,15])
cor=np.corrcoef(data, data2)

print("correlation coefficient: ",cor)
```

```python
#save data into a file

np.savetxt('matrix_output.csv',mul, delimiter=',')
```

```python
#save output into a binary file

np.save('matrix_output.npy', mul)
new_matrix = np.load('matrix_output.npy')
#note: .npy is NumPy binary file format
```
